```python
# consolidated_with_rf.py
# Additions: sampling, profile-based botnet labeling, pseudo-label RF classifier
import os
import time
import json
import warnings
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import boxcox
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from sklearn.metrics import silhouette_score
import hdbscan

sns.set(style="whitegrid")
warnings.filterwarnings("ignore")
```

```python
# -------------------------
# CONFIG
# -------------------------
TRAIN_PATH = r'X:\Dissertacao\python_projects\dataset\ISCX-Bot-2014\ISCX_csv\Testing_file.csv' 
TEST_PATH  = r'X:\Dissertacao\python_projects\dataset\ISCX-Bot-2014\ISCX_csv\Training_file.csv'    
SAMPLE_SIZE = 200_000              # sample size for experiments
RANDOM_STATE = 42
PCA_COMPONENTS = 10
HDB_MIN_CLUSTER = 200
HDB_MIN_SAMPLES = 20
IFOREST_PARAMS = dict(n_estimators=400, contamination=0.2, random_state=RANDOM_STATE)
RF_PARAMS = dict(n_estimators=400, random_state=RANDOM_STATE, n_jobs=-1)
NUM_FEATURES = [
    "Time_Diff","Log_IATime","Log_BRate",
    "BoxCox_Length","BoxCox_PRate",
    "Length_Mean","Length_Std","Pkt_Per_Src",
    "Session_Dur_Src","Session_Dur_Dst"
]
```

```python
# -------------------------
# 1. Load (small wrapper, adjust encoding if needed)
# -------------------------
print("Loading datasets...")
df_train = pd.read_csv(TRAIN_PATH, encoding='ISO-8859-1', low_memory=False)
df_test = pd.read_csv(TEST_PATH, encoding='ISO-8859-1', low_memory=False)
```

```python
# -------------------------
# 2. Basic cleaning & ensure minimal columns
# -------------------------
for df in (df_train, df_test):
    if 'Info' in df.columns:
        df['Info'] = df['Info'].fillna("Unknown")
    if 'Time' in df.columns:
        df = df[df['Time'].notna()]
    # remove rows with missing Source/Destination
df_train.dropna(subset=['Source','Destination'], inplace=True)
df_test.dropna(subset=['Source','Destination'], inplace=True)
```

```python
# -------------------------
# 3. Safe categorical encoding (protocol, source, destination) using combined mapping
# -------------------------
def safe_map_cols(train, test, col):
    values = pd.concat([train[col].astype(str), test[col].astype(str)], axis=0).unique()
    mapping = {v: i for i, v in enumerate(values)}
    train[col + "_enc"] = train[col].astype(str).map(mapping).fillna(-1).astype(int)
    test[col + "_enc"]  = test[col].astype(str).map(mapping).fillna(-1).astype(int)
    return mapping

print("Encoding categorical columns...")
protocol_map = safe_map_cols(df_train, df_test, 'Protocol') if 'Protocol' in df_train.columns else {}
source_map = safe_map_cols(df_train, df_test, 'Source')
dest_map = safe_map_cols(df_train, df_test, 'Destination')

# rename enc columns expected by feature builder
df_train.rename(columns={'Source_enc': 'Source_enc', 'Destination_enc': 'Destination_enc'}, inplace=True)
df_test.rename(columns={'Source_enc': 'Source_enc', 'Destination_enc': 'Destination_enc'}, inplace=True)

# for consistency create Source_enc / Destination_enc if not present
if 'Source_enc' not in df_train.columns: df_train['Source_enc'] = df_train['Source'].astype(str).map(source_map).fillna(-1).astype(int)
if 'Source_enc' not in df_test.columns: df_test['Source_enc']  = df_test['Source'].astype(str).map(source_map).fillna(-1).astype(int)
if 'Destination_enc' not in df_train.columns: df_train['Destination_enc'] = df_train['Destination'].astype(str).map(dest_map).fillna(-1).astype(int)
if 'Destination_enc' not in df_test.columns: df_test['Destination_enc']  = df_test['Destination'].astype(str).map(dest_map).fillna(-1).astype(int)
```

```python
# -------------------------
# 4. Feature engineering (robust, same as before but encapsulated)
# -------------------------
def build_features_safe(df):
    df = df.copy()
    src_col = 'Source_enc' if 'Source_enc' in df.columns else 'Source'
    dst_col = 'Destination_enc' if 'Destination_enc' in df.columns else 'Destination'

    # Time_Diff per-source
    df['Time_Diff'] = df.groupby(src_col)['Time'].diff()
    df['Time_Diff'] = df['Time_Diff'].fillna(df.groupby(src_col)['Time'].transform('median'))
    df['Time_Diff'] = df['Time_Diff'].fillna(0.0)

    # Packet_Rate robust
    def packet_rate(series):
        arr = np.asarray(series.dropna())
        if arr.size < 2:
            return 0.0
        q1, q3 = np.percentile(arr, [25, 75])
        iqr = max(q3 - q1, 1e-9)
        clipped = np.clip(arr, q1 - 1.5*iqr, q3 + 1.5*iqr)
        denom = clipped.max() - clipped.min()
        if denom <= 1e-6:
            return 0.0
        return clipped.size / denom

    df['Packet_Rate'] = df.groupby(src_col)['Time'].transform(packet_rate)
    df['Packet_Rate'] = df['Packet_Rate'].replace([np.inf, -np.inf], 0).fillna(0.0)

    # Inter-arrival rolling mean
    df['Inter_Arrival_Time'] = df.groupby(src_col)['Time_Diff'].transform(lambda s: s.rolling(10, min_periods=1).mean())
    df['Inter_Arrival_Time'] = df['Inter_Arrival_Time'].clip(lower=1e-6)

    # Burst rate
    df['Burst_Rate'] = np.where(df['Inter_Arrival_Time'] > 1e-6, 1.0 / df['Inter_Arrival_Time'], 0.0)

    # Variability features
    df['Length_Mean'] = df.groupby(src_col)['Length'].transform('mean').fillna(df['Length'].mean())
    df['Length_Std']  = df.groupby(src_col)['Length'].transform('std').fillna(0.0)
    df['Pkt_Per_Src'] = df.groupby(src_col)['Length'].transform('count').fillna(0).astype(int)

    # Session durations per source/dest
    df['Session_Dur_Src'] = df.groupby(src_col)['Time'].transform(lambda x: x.max() - x.min())
    df['Session_Dur_Dst'] = df.groupby(dst_col)['Time'].transform(lambda x: x.max() - x.min())

    # Transforms
    df['Log_IATime'] = np.log1p(df['Inter_Arrival_Time'])
    df['Log_BRate']  = np.log1p(df['Burst_Rate'].clip(lower=0))

    # BoxCox with safe fallbacks
    def safe_boxcox(series, name):
        arr = series.fillna(0.0).astype(float) + 1e-6
        try:
            out, lam = boxcox(arr + 1e-6)
            return out
        except Exception:
            return np.log1p(arr)

    df['BoxCox_Length'] = safe_boxcox(df['Length'], "Length")
    df['BoxCox_PRate'] = safe_boxcox(df['Packet_Rate'], "Packet_Rate")

    # Ensure numeric columns are clean
    for c in ['Time_Diff','Packet_Rate','Inter_Arrival_Time','Burst_Rate','Log_IATime','Log_BRate',
              'Length_Mean','Length_Std','Pkt_Per_Src','Session_Dur_Src','Session_Dur_Dst',
              'BoxCox_Length','BoxCox_PRate']:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0.0)

    return df

print("Building features (train/test)...")
df_train = build_features_safe(df_train)
df_test = build_features_safe(df_test)
```

```python
# -------------------------
# 5. Sample for experiments
# -------------------------
print(f"Sampling {SAMPLE_SIZE} rows for experiments (train/test)...")
def sample_with_anoms(df, n=SAMPLE_SIZE, seed=RANDOM_STATE):
    if n >= len(df): 
        return df.sample(frac=1.0, random_state=seed).reset_index(drop=True)
    # keep a slice with some anomalies (we don't have labels yet; random sample)
    return df.sample(n=n, random_state=seed).reset_index(drop=True)

df_train_sample = sample_with_anoms(df_train, n=SAMPLE_SIZE)
df_test_sample = sample_with_anoms(df_test, n=SAMPLE_SIZE)
```

```python
# -------------------------
# 6. Scaling (RobustScaler)
# -------------------------
print("Scaling numeric features with RobustScaler...")
scaler = RobustScaler()
scaler.fit(df_train_sample[NUM_FEATURES])
X_train = pd.DataFrame(scaler.transform(df_train_sample[NUM_FEATURES]), columns=NUM_FEATURES)
X_test = pd.DataFrame(scaler.transform(df_test_sample[NUM_FEATURES]), columns=NUM_FEATURES)

# Keep indices aligned for mapping back
X_train.index = df_train_sample.index
X_test.index = df_test_sample.index
```

```python
# -------------------------
# 7. Isolation Forest (train on sample)
# -------------------------
print("Training Isolation Forest on sample...")
iso = IsolationForest(**IFOREST_PARAMS)
t0 = time.time()
iso.fit(X_train)
t1 = time.time()
print(f"IsolationForest trained in {(t1-t0):.1f}s")

df_train_sample['Anomaly_IForest'] = iso.predict(X_train)   # -1 anomaly, 1 normal
df_test_sample['Anomaly_IForest'] = iso.predict(X_test)
```

```python
# -------------------------
# 8. PCA fit once on training sample, then HDBSCAN on PCA embedding
# -------------------------
print("Fitting PCA and running HDBSCAN on training sample...")
pca = PCA(n_components=PCA_COMPONENTS, random_state=RANDOM_STATE)
X_train_pca = pca.fit_transform(X_train)

clusterer = hdbscan.HDBSCAN(min_cluster_size=HDB_MIN_CLUSTER, min_samples=HDB_MIN_SAMPLES,
                            metric='euclidean', cluster_selection_method='eom', prediction_data=True)
t0 = time.time()
cluster_labels = clusterer.fit_predict(X_train_pca)   # -1 = noise/anomaly
t1 = time.time()
print(f"HDBSCAN fit in {(t1-t0):.1f}s; clusters found: {len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)}, noise fraction: {(cluster_labels==-1).mean():.4f}")

df_train_sample['HDBSCAN_Label'] = cluster_labels
df_train_sample['HDBSCAN_IsNoise'] = (cluster_labels == -1).astype(int)

# Project test sample using same PCA and approximate_predict
X_test_pca = pca.transform(X_test)
labels_test, strengths_test = hdbscan.approximate_predict(clusterer, X_test_pca)
df_test_sample['HDBSCAN_Label'] = labels_test
df_test_sample['HDBSCAN_IsNoise'] = (labels_test == -1).astype(int)
```

```python
# -------------------------
# 9. Pseudo-label creation (consensus) & silhouette
# -------------------------
# Consensus anomaly = both IForest says -1 and HDBSCAN marks noise
df_train_sample['Consensus_Anomaly'] = ((df_train_sample['Anomaly_IForest'] == -1) & (df_train_sample['HDBSCAN_IsNoise'] == 1)).astype(int)
df_test_sample['Consensus_Anomaly'] = ((df_test_sample['Anomaly_IForest'] == -1) & (df_test_sample['HDBSCAN_IsNoise'] == 1)).astype(int)

# Convert to conventions used earlier (-1 for anomaly)
df_train_sample['Anomaly_IForest_bin'] = (df_train_sample['Anomaly_IForest'] == -1).astype(int)
df_test_sample['Anomaly_IForest_bin'] = (df_test_sample['Anomaly_IForest'] == -1).astype(int)

# Silhouette on the sample (IForest labels & HDBSCAN noise labels)
def safe_silhouette(X_df, labels):
    try:
        if len(np.unique(labels)) < 2:
            return np.nan
        return silhouette_score(X_df[NUM_FEATURES], labels)
    except Exception:
        return np.nan

sil_if_train = safe_silhouette(X_train, df_train_sample['Anomaly_IForest'])
sil_hdb_train = safe_silhouette(X_train, df_train_sample['HDBSCAN_IsNoise'])
print(f"Silhouette IForest (sample train): {sil_if_train:.4f}")
print(f"Silhouette HDBSCAN (sample train): {sil_hdb_train:.4f}")
```

```python
# -------------------------
# 10. Behavioral profiles -> label anomalies with probable botnet types
# -------------------------
# THESE RULES ARE HEURISTICS â€” tweak ranges to fit your dataset
botnet_profiles = {
    "HighBurst": {"Pkt_Per_Src": (500, np.inf), "Burst_Rate": (5.0, np.inf)},   # very bursty sources
    "Sustained_HighRate": {"Pkt_Per_Src": (200, 500), "Inter_Arrival_Time": (0.0, 0.05)},
    "Short_Bursts": {"Burst_Rate": (2.0, 5.0), "Session_Dur_Src": (0.0, 300.0)},
    "Large_Packets": {"BoxCox_Length": (np.percentile(df_train_sample['BoxCox_Length'], 75), np.inf)},
}

def match_profile(row, profiles):
    for name, rules in profiles.items():
        ok = True
        for feat, (low, high) in rules.items():
            if feat not in row or pd.isna(row[feat]):
                ok = False; break
            val = row[feat]
            if not (val >= low and val <= high):
                ok = False; break
        if ok:
            return name
    return "Unknown"

# Apply to anomalies only (use original numeric columns in sample frames)
for df_s in [df_train_sample, df_test_sample]:
    df_s['Pred_Botnet_Profile'] = "Normal"
    mask = df_s['Consensus_Anomaly'] == 1
    df_s.loc[mask, 'Pred_Botnet_Profile'] = df_s[mask].apply(lambda r: match_profile(r, botnet_profiles), axis=1)
```

```python
# -------------------------
# 11. Prepare supervised labels for RandomForest
#    We'll create a multiclass label: 'Normal' and profile names where matched.
# -------------------------
def prepare_rf_dataset(df_s):
    # Use rows where Pred_Botnet_Profile != "Unknown" or treat all as classes (Unknown included)
    # Label = Pred_Botnet_Profile, but normal rows labeled "Normal"
    df_copy = df_s.copy()
    df_copy['RF_Label'] = df_copy['Pred_Botnet_Profile'].fillna("Unknown")
    # If you want to collapse rare classes, do it here (not done now)
    return df_copy

rf_train_df = prepare_rf_dataset(df_train_sample)
rf_test_df = prepare_rf_dataset(df_test_sample)

# For RF training, to reduce label noise, we'll train on:
# - all "Normal" rows
# - and anomalies that have a specific profile (not "Unknown")
train_mask = (rf_train_df['RF_Label'] == "Normal") | (rf_train_df['RF_Label'] != "Unknown")
rf_train_sub = rf_train_df[train_mask].copy()
print("RF training set distribution:", Counter(rf_train_sub['RF_Label']))

# If there are too many normals, downsample normals to balance
class_counts = rf_train_sub['RF_Label'].value_counts()
min_class = class_counts[class_counts.index != "Normal"].min() if any(class_counts.index != "Normal") else class_counts.min()
if 'Normal' in class_counts and min_class > 0:
    normals = rf_train_sub[rf_train_sub['RF_Label'] == "Normal"]
    others = rf_train_sub[rf_train_sub['RF_Label'] != "Normal"]
    n_normals = min(len(normals), int(min_class * len(others['RF_Label'].unique()) * 3))  # keep some majority
    normals_sampled = normals.sample(n=n_normals, random_state=RANDOM_STATE)
    rf_train_bal = pd.concat([normals_sampled, others]).sample(frac=1.0, random_state=RANDOM_STATE)
else:
    rf_train_bal = rf_train_sub.copy()

print("Final RF training distribution:", Counter(rf_train_bal['RF_Label']))
```

```python
# -------------------------
# 12. Train RandomForest classifier (multi-class)
# -------------------------
print("Training RandomForest (pseudo-labels)...")
rf = RandomForestClassifier(**RF_PARAMS)
rf.fit(rf_train_bal[NUM_FEATURES], rf_train_bal['RF_Label'])

# Predict on test sample
y_true = rf_test_df['RF_Label']
y_pred = rf.predict(rf_test_df[NUM_FEATURES])

print("RandomForest classification report (test sample):")
print(classification_report(y_true, y_pred, zero_division=0))

# Feature importances
feat_imp = pd.Series(rf.feature_importances_, index=NUM_FEATURES).sort_values(ascending=False)
```

```python
# -------------------------
# 13. Summary statistics & evaluation
# -------------------------
def anomaly_pct(df, col):
    return (df[col] == 1).sum() / len(df) * 100

if_pct_train = (df_train_sample['Anomaly_IForest'] == -1).mean() * 100
if_pct_test = (df_test_sample['Anomaly_IForest'] == -1).mean() * 100
hdb_pct_train = (df_train_sample['HDBSCAN_IsNoise'] == 1).mean() * 100
hdb_pct_test = (df_test_sample['HDBSCAN_IsNoise'] == 1).mean() * 100
cons_pct_train = df_train_sample['Consensus_Anomaly'].mean() * 100
cons_pct_test = df_test_sample['Consensus_Anomaly'].mean() * 100

agreement_train = ( (df_train_sample['Anomaly_IForest'] == -1).astype(int) == df_train_sample['HDBSCAN_IsNoise']).mean() * 100
agreement_test = ( (df_test_sample['Anomaly_IForest'] == -1).astype(int) == df_test_sample['HDBSCAN_IsNoise']).mean() * 100

summary = {
    "IForest_train_pct": round(if_pct_train,2),
    "IForest_test_pct": round(if_pct_test,2),
    "HDBSCAN_train_pct": round(hdb_pct_train,2),
    "HDBSCAN_test_pct": round(hdb_pct_test,2),
    "Consensus_train_pct": round(cons_pct_train,2),
    "Consensus_test_pct": round(cons_pct_test,2),
    "Agreement_train_pct": round(agreement_train,2),
    "Agreement_test_pct": round(agreement_test,2),
    "Silhouette_IF_train": round(sil_if_train,4),
    "Silhouette_HDB_train": round(sil_hdb_train,4)
}

print("=== SUMMARY ===")
print(json.dumps(summary, indent=2))

# Top predicted botnets in test
top_botnets_test = df_test_sample['Pred_Botnet_Profile'].value_counts().head(10)
print("Top predicted botnet profiles (test sample):")
print(top_botnets_test)
```

```python
# -------------------------
# 14. Plots: Feature importance + few KDEs
# -------------------------
plt.figure(figsize=(8,5))
feat_imp.plot(kind='barh')
plt.title("RandomForest feature importances (pseudo-labels)")
plt.tight_layout()
plt.show()

# KDE plots for a subset of features (shows Normal vs Anomaly by consensus)
kde_features = ["Pkt_Per_Src","Burst_Rate","Inter_Arrival_Time","BoxCox_Length"]
for f in kde_features:
    plt.figure(figsize=(7,4))
    sns.kdeplot(df_test_sample.loc[df_test_sample['Consensus_Anomaly']==0, f], label="Normal (consensus)", fill=True)
    sns.kdeplot(df_test_sample.loc[df_test_sample['Consensus_Anomaly']==1, f], label="Consensus Anom", fill=True)
    plt.title(f"KDE Test Sample: {f}")
    plt.legend()
    plt.show()

# Confusion matrix for RF multiclass (test)
cm = confusion_matrix(y_true, y_pred, labels=rf.classes_)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=rf.classes_, yticklabels=rf.classes_, cmap="Blues")
plt.title("RandomForest confusion matrix (test sample)")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.show()
```

```python
# -------------------------
# 15. Persist summary to disk
# -------------------------
out = {
    "summary": summary,
    "rf_feature_importances": feat_imp.to_dict(),
    "top_botnets_test": top_botnets_test.to_dict()
}
with open("detection_summary.json", "w") as f:
    json.dump(out, f, indent=2)

print("Pipeline complete. Summary saved to detection_summary.json")
```
