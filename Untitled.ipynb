```python
import time
import json
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox

from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, silhouette_score
import hdbscan

from imblearn.over_sampling import SMOTE

sns.set(style="whitegrid")
```

```python
# -------------------------
# CONFIG
# -------------------------
TRAIN_PATH = r'X:\Dissertacao\python_projects\dataset\ISCX-Bot-2014\ISCX_csv\Testing_file.csv' 
TEST_PATH  = r'X:\Dissertacao\python_projects\dataset\ISCX-Bot-2014\ISCX_csv\Training_file.csv'    
SAMPLE_SIZE = 175_000              # sample size for experiments
RANDOM_STATE = 42

# Model / transform hyperparams (tuned suggestions)
HDB_MIN_CLUSTER = 1200  # double the previous(600)
HDB_MIN_SAMPLES = 25
PCA_COMPONENTS = 3
IFOREST_PARAMS = dict(n_estimators=250, contamination=0.05, max_samples=0.5, random_state=RANDOM_STATE)
RF_PARAMS = dict(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1, class_weight='balanced_subsample')

# Features to use for modeling (ensure these are produced by build_features)
NUM_FEATURES = [
    "Time_Diff","Log_IATime","Log_BRate",
    "BoxCox_Length","BoxCox_PRate",
    "Length_Mean","Length_Std","Pkt_Per_Src",
    "Session_Dur_Src","Session_Dur_Dst"
]

# -------------------------
# Utility functions
# -------------------------
def safe_boxcox(series):
    arr = np.asarray(series.fillna(0.0).astype(float) + 1e-6)
    try:
        out, _ = boxcox(arr + 1e-6)
        return out
    except Exception:
        return np.log1p(arr)

def safe_sample(df, n, seed=RANDOM_STATE):
    if n >= len(df):
        return df.sample(frac=1.0, random_state=seed).reset_index(drop=True)
    return df.sample(n=n, random_state=seed).reset_index(drop=True)
```

```python
# -------------------------
# 1. Load (small wrapper, adjust encoding if needed)
# -------------------------
print("Loading datasets...")
df_train = pd.read_csv(TRAIN_PATH, encoding='ISO-8859-1', low_memory=False)
df_test = pd.read_csv(TEST_PATH, encoding='ISO-8859-1', low_memory=False)
```

```python
# Basic minimal cleaning & ensure necessary columns exist
for df in (df_train, df_test):
    if 'Info' in df.columns:
        df['Info'] = df['Info'].fillna("Unknown")
    # drop rows with missing Source/Destination/time/length if present
    for c in ['Source','Destination','Time','Length']:
        if c in df.columns:
            df.dropna(subset=[c], inplace=True)
```

```python
# -------------------------
# 2. Safe combined encoding for categorical columns (Protocol/Source/Destination)
# -------------------------
def safe_map(train, test, col):
    # map union of unique values to int ids (keeps same mapping across train/test)
    uvals = pd.concat([train[col].astype(str), test[col].astype(str)], axis=0).unique()
    mapping = {v: i for i, v in enumerate(uvals)}
    train[col + "_enc"] = train[col].astype(str).map(mapping).fillna(-1).astype(int)
    test[col + "_enc"]  = test[col].astype(str).map(mapping).fillna(-1).astype(int)
    return mapping

print("Encoding categorical columns (safe)...")
if 'Protocol' in df_train.columns:
    _ = safe_map(df_train, df_test, 'Protocol')
_ = safe_map(df_train, df_test, 'Source')
_ = safe_map(df_train, df_test, 'Destination')

# create Source_enc / Destination_enc if missing
if 'Source_enc' not in df_train.columns and 'Source_enc' in df_train.columns:
    df_train['Source_enc'] = df_train['Source'].astype(str).map(_).fillna(-1).astype(int)
if 'Source_enc' not in df_train.columns:
    df_train['Source_enc'] = df_train['Source'].astype(str).astype('category').cat.codes
if 'Source_enc' not in df_test.columns:
    df_test['Source_enc']  = df_test['Source'].astype(str).map(pd.Series(df_train['Source_enc'].values, index=df_train['Source'].astype(str))).fillna(-1).astype(int) if 'Source' in df_test.columns else -1

if 'Destination_enc' not in df_train.columns:
    df_train['Destination_enc'] = df_train['Destination'].astype(str).astype('category').cat.codes
if 'Destination_enc' not in df_test.columns:
    df_test['Destination_enc']  = df_test['Destination'].astype(str).map(pd.Series(df_train['Destination_enc'].values, index=df_train['Destination'].astype(str))).fillna(-1).astype(int) if 'Destination' in df_test.columns else -1
```

```python
# -------------------------
# 3. Feature engineering (robust)
# -------------------------
def build_features(df):
    df = df.copy()
    src = 'Source_enc' if 'Source_enc' in df.columns else 'Source'
    dst = 'Destination_enc' if 'Destination_enc' in df.columns else 'Destination'

    # Time diff per source
    if 'Time' in df.columns:
        df['Time_Diff'] = df.groupby(src)['Time'].diff()
        df['Time_Diff'] = df['Time_Diff'].fillna(df.groupby(src)['Time'].transform('median'))
        df['Time_Diff'] = df['Time_Diff'].fillna(0.0)
    else:
        df['Time_Diff'] = 0.0

    # Packet rate per source (robust IQR clipping approach)
    def pkt_rate(series):
        a = np.asarray(series.dropna())
        if a.size < 2:
            return 0.0
        q1, q3 = np.percentile(a, [25, 75])
        iqr = max(q3 - q1, 1e-9)
        cl = np.clip(a, q1 - 1.5*iqr, q3 + 1.5*iqr)
        denom = cl.max() - cl.min()
        if denom <= 1e-6:
            return 0.0
        return cl.size / denom

    if 'Time' in df.columns:
        df['Packet_Rate'] = df.groupby(src)['Time'].transform(pkt_rate)
    else:
        df['Packet_Rate'] = 0.0
    df['Packet_Rate'] = df['Packet_Rate'].replace([np.inf,-np.inf], 0.0).fillna(0.0)

    # rolling inter-arrival, burst rate
    df['Inter_Arrival_Time'] = df.groupby(src)['Time_Diff'].transform(lambda s: s.rolling(10, min_periods=1).mean()).fillna(0.0)
    df['Inter_Arrival_Time'] = df['Inter_Arrival_Time'].clip(lower=1e-6)
    df['Burst_Rate'] = np.where(df['Inter_Arrival_Time'] > 1e-6, 1.0 / df['Inter_Arrival_Time'], 0.0)

    # variability features
    df['Length_Mean'] = df.groupby(src)['Length'].transform('mean').fillna(df['Length'].mean() if 'Length' in df.columns else 0.0)
    df['Length_Std']  = df.groupby(src)['Length'].transform('std').fillna(0.0)
    df['Pkt_Per_Src'] = df.groupby(src)['Length'].transform('count').fillna(0).astype(int)

    # session durations
    if 'Time' in df.columns:
        df['Session_Dur_Src'] = df.groupby(src)['Time'].transform(lambda x: x.max() - x.min())
        df['Session_Dur_Dst'] = df.groupby(dst)['Time'].transform(lambda x: x.max() - x.min())
    else:
        df['Session_Dur_Src'] = 0.0
        df['Session_Dur_Dst'] = 0.0

    # transforms: safe logs/boxcox
    df['Log_IATime'] = np.log1p(df['Inter_Arrival_Time'])
    df['Log_BRate']  = np.log1p(df['Burst_Rate'].clip(lower=0))

    df['BoxCox_Length'] = safe_boxcox(df['Length']) if 'Length' in df.columns else np.zeros(len(df))
    df['BoxCox_PRate']  = safe_boxcox(df['Packet_Rate'])

    # ensure numeric columns exist and are finite
    for c in NUM_FEATURES:
        if c not in df.columns:
            df[c] = 0.0
        df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0.0)

    return df

print("Building features for train/test (may take some time)...")
df_train = build_features(df_train)
df_test  = build_features(df_test)
```

```python
# -------------------------
# 4. Sampling for experiment
# -------------------------
print(f"Sampling up to {SAMPLE_SIZE} rows from train/test for experiments...")
df_train_sample = safe_sample(df_train, SAMPLE_SIZE)
df_test_sample  = safe_sample(df_test, SAMPLE_SIZE)
```

```python
# -------------------------
# 5. Scaling (fit on train sample, reuse for test)
# -------------------------
print("Scaling numeric features with RobustScaler (fit on train sample)...")
scaler = RobustScaler()
scaler.fit(df_train_sample[NUM_FEATURES])

X_train = pd.DataFrame(scaler.transform(df_train_sample[NUM_FEATURES]), columns=NUM_FEATURES, index=df_train_sample.index)
X_test  = pd.DataFrame(scaler.transform(df_test_sample[NUM_FEATURES]), columns=NUM_FEATURES, index=df_test_sample.index)
```

```python
# -------------------------
# 6. Isolation Forest (train on sample)
# -------------------------
print("Training IsolationForest on train sample...")
iso = IsolationForest(**IFOREST_PARAMS)
t0 = time.time()
iso.fit(X_train)
t1 = time.time()
print(f"IsolationForest trained in {(t1-t0):.1f}s")

df_train_sample['Anomaly_IForest'] = iso.predict(X_train)   # -1 = anomaly
df_test_sample['Anomaly_IForest']  = iso.predict(X_test)
```

```python
# -------------------------
# 7. PCA fit once (train) and HDBSCAN on PCA embedding (train sample)
# -------------------------
print(f"Fitting PCA (n_components={PCA_COMPONENTS}) and HDBSCAN (min_cluster_size={HDB_MIN_CLUSTER}) on train sample...")
pca = PCA(n_components=PCA_COMPONENTS, random_state=RANDOM_STATE)
X_train_pca = pca.fit_transform(X_train)

clusterer = hdbscan.HDBSCAN(min_cluster_size=HDB_MIN_CLUSTER,
                            min_samples=HDB_MIN_SAMPLES,
                            metric='euclidean',
                            cluster_selection_method='eom',
                            prediction_data=True)

t0 = time.time()
train_labels = clusterer.fit_predict(X_train_pca)   # -1 = noise
t1 = time.time()
print(f"HDBSCAN fit in {(t1-t0):.1f}s. clusters found (excluding noise): {len(set(train_labels)) - (1 if -1 in train_labels else 0)}")
df_train_sample['HDBSCAN_Label'] = train_labels
df_train_sample['HDBSCAN_IsNoise'] = (train_labels == -1).astype(int)

# Use approximate_predict to project labels to the test sample
X_test_pca = pca.transform(X_test)
test_labels, _ = hdbscan.approximate_predict(clusterer, X_test_pca)
df_test_sample['HDBSCAN_Label'] = test_labels
df_test_sample['HDBSCAN_IsNoise'] = (test_labels == -1).astype(int)
```

```python
# -------------------------
# 8. Consensus pseudo-labels for supervised RF
# -------------------------
# consensus anomaly: both IF and HDBSCAN mark it anomalous
df_train_sample['Consensus_Anomaly'] = ((df_train_sample['Anomaly_IForest'] == -1) & (df_train_sample['HDBSCAN_IsNoise'] == 1)).astype(int)
df_test_sample['Consensus_Anomaly']  = ((df_test_sample['Anomaly_IForest'] == -1) & (df_test_sample['HDBSCAN_IsNoise'] == 1)).astype(int)
```

```python
# -------------------------
# 9. Botnet behavior profiles (heuristic rules) - tune these ranges to your dataset
# -------------------------
percentiles = [10, 30, 50, 70, 80, 90, 95]
p = {}  # p[feature] = {perc: value}
features_for_percentiles = ["Pkt_Per_Src", "Burst_Rate", "Session_Dur_Src",
                            "Inter_Arrival_Time", "BoxCox_Length", "Pkt_Per_Src"]

for feat in features_for_percentiles:
    if feat in df_train_sample.columns:
        vals = df_train_sample[feat].dropna().values
        if vals.size == 0:
            # fallback to small non-zero range
            p[feat] = {perc: 0.0 for perc in percentiles}
        else:
            qvals = np.percentile(vals, percentiles)
            p[feat] = {perc: float(q) for perc, q in zip(percentiles, qvals)}
    else:
        # feature missing -> zeroed percentiles
        p[feat] = {perc: 0.0 for perc in percentiles}

# Helpful shorthand for grabbing percentiles (safe)
def pp(feat, perc):
    return p.get(feat, {}).get(perc, 0.0)

botnet_profiles = {
    "DDoS": {
        # very high packet counts (>= 95th), very high burst rate (>= 90th), short sessions (<= 30th)
        "Pkt_Per_Src": (pp("Pkt_Per_Src", 95), np.inf),
        "Burst_Rate": (pp("Burst_Rate", 90), np.inf),
        "Session_Dur_Src": (0.0, pp("Session_Dur_Src", 30))
    },
    "SPAM": {
        # moderately high packet counts (70-95th), small packet sizes (BoxCox_Length <= 30th)
        "Pkt_Per_Src": (pp("Pkt_Per_Src", 70), pp("Pkt_Per_Src", 95)),
        "BoxCox_Length": ( -np.inf, pp("BoxCox_Length", 30) )
    },
    "ClickFraud": {
        # mid-high burstiness and small IAT (between 60th and 90th for burst, low IAT)
        "Burst_Rate": (pp("Burst_Rate", 60), pp("Burst_Rate", 90)),
        "Inter_Arrival_Time": (0.0, pp("Inter_Arrival_Time", 30))
    },
    "PortScan": {
        # high packet counts per src (>= 80th), shortish sessions (<= median)
        "Pkt_Per_Src": (pp("Pkt_Per_Src", 80), np.inf),
        "Session_Dur_Src": (0.0, pp("Session_Dur_Src", 50))
    }
}

# For debugging: print the percentile map and the profile rules
print("Percentiles (sample):")
for feat in features_for_percentiles:
    print(f" {feat}: ", {perc: p[feat][perc] for perc in percentiles})
print("\nConstructed botnet_profiles (percentile-driven):")
for name, rules in botnet_profiles.items():
    print(f" {name}:")
    for k, (low, high) in rules.items():
        print(f"   {k}: {low:.6g} -> {high:.6g}")

def match_profile_row(row, profiles):
    for name, rules in profiles.items():
        ok = True
        for feat, (low, high) in rules.items():
            if feat not in row or pd.isna(row[feat]):
                ok = False
                break
            val = row[feat]
            if not (val >= low and val <= high):
                ok = False
                break
        if ok:
            return name
    return "Unknown"

# Apply profile matching only to consensus anomalies (reduces false labelling)
for df_s in [df_train_sample, df_test_sample]:
    df_s['Pred_Botnet_Profile'] = "Normal"
    mask = df_s['Consensus_Anomaly'] == 1
    if mask.any():
        df_s.loc[mask, 'Pred_Botnet_Profile'] = df_s[mask].apply(lambda r: match_profile_row(r, botnet_profiles), axis=1)
```

```python
# -------------------------
# 10. Prepare RF dataset from pseudo-labels
#    Labels: 'Normal' or profile name (filter Unknown if you want)
# -------------------------
def prepare_rf_df(df_s):
    dfc = df_s.copy()
    dfc['RF_Label'] = dfc['Pred_Botnet_Profile'].fillna("Unknown")
    # Option: treat "Unknown" as a separate class or drop them for RF training to reduce label noise
    return dfc

rf_train_df = prepare_rf_df(df_train_sample)
rf_test_df  = prepare_rf_df(df_test_sample)

# Build training set: include all consensus anomalies with known profile + a balanced sample of Normal
anom_known = rf_train_df[rf_train_df['RF_Label'] != "Unknown"]
normals = rf_train_df[rf_train_df['RF_Label'] == "Normal"]

# If there are very few anomalies with known profile, we'll keep them as-is; otherwise balance
if len(anom_known) == 0:
    print("No consensus anomalies with known profile found - RF will not be trained.")
    rf_train_ready = rf_train_df.sample(min(10000, len(rf_train_df)), random_state=RANDOM_STATE)  # fallback small sample
else:
    # target: keep a multiple of the largest non-normal class per class to balance better
    class_counts = anom_known['RF_Label'].value_counts()
    min_non_norm = class_counts.min()
    # sample some normals to keep class balance reasonable
    n_normals_keep = min(len(normals), int(min_non_norm * max(5, len(class_counts))))
    normals_sampled = normals.sample(n=n_normals_keep, random_state=RANDOM_STATE) if len(normals)>0 else pd.DataFrame()
    rf_train_ready = pd.concat([anom_known, normals_sampled]).sample(frac=1.0, random_state=RANDOM_STATE).reset_index(drop=True)

rf_train_ready = rf_train_ready[rf_train_ready['RF_Label'] != "Unknown"]
print("RF training class distribution:", Counter(rf_train_ready['RF_Label']))
```

```python
# -------------------------
# 11. Train RandomForest (if we have classes)
# -------------------------
if 'RF_Label' in rf_train_ready.columns and len(rf_train_ready) > 10:
    print("Training RandomForest classifier on pseudo-labels (balanced_subsample)...")
    rf = RandomForestClassifier(**RF_PARAMS)
    sm = SMOTE(random_state=42)
    X_bal, y_bal = sm.fit_resample(rf_train_ready[NUM_FEATURES], rf_train_ready['RF_Label'])
    rf.fit(X_bal, y_bal)
    
    # Predict on test sample
    y_true = rf_test_df['RF_Label']
    y_pred = rf.predict(rf_test_df[NUM_FEATURES])
    print("RandomForest classification report (test sample):")
    print(classification_report(y_true, y_pred, zero_division=0))
    feat_imp = pd.Series(rf.feature_importances_, index=NUM_FEATURES).sort_values(ascending=False)
else:
    rf = None
    feat_imp = pd.Series(0, index=NUM_FEATURES)
```

```python
# -------------------------
# 12. Metrics & summary
# -------------------------
def pct(col):
    return round(100.0 * np.mean(col), 2)

summary = {
    "IForest_train_pct": pct(df_train_sample['Anomaly_IForest'] == -1),
    "IForest_test_pct": pct(df_test_sample['Anomaly_IForest'] == -1),
    "HDBSCAN_train_pct": pct(df_train_sample['HDBSCAN_IsNoise'] == 1),
    "HDBSCAN_test_pct": pct(df_test_sample['HDBSCAN_IsNoise'] == 1),
    "Consensus_train_pct": pct(df_train_sample['Consensus_Anomaly'] == 1),
    "Consensus_test_pct": pct(df_test_sample['Consensus_Anomaly'] == 1),
    "Agreement_train_pct": pct((df_train_sample['Anomaly_IForest'] == -1).astype(int) == df_train_sample['HDBSCAN_IsNoise']),
    "Agreement_test_pct": pct((df_test_sample['Anomaly_IForest'] == -1).astype(int) == df_test_sample['HDBSCAN_IsNoise']),
    "Silhouette_IF_train": round(silhouette_score(X_train, (df_train_sample['Anomaly_IForest'] == -1).astype(int)) if len(np.unique(df_train_sample['Anomaly_IForest']))>1 else float('nan'), 4),
    "Silhouette_HDB_train": round(silhouette_score(X_train, df_train_sample['HDBSCAN_IsNoise']) if len(np.unique(df_train_sample['HDBSCAN_IsNoise']))>1 else float('nan'), 4)
}

print("=== SUMMARY ===")
print(json.dumps(summary, indent=2))

# Top predicted botnets in test
print("Top predicted botnet profiles (test sample):")
print(df_test_sample['Pred_Botnet_Profile'].value_counts().head(10))
```

```python
# -------------------------
# 13. Plots: feature importances + KDEs + confusion matrix
# -------------------------
if rf is not None:
    plt.figure(figsize=(8,5))
    feat_imp.plot(kind='barh')
    plt.title("RandomForest feature importances (pseudo-labels)")
    plt.tight_layout()
    plt.show()

# KDEs for key features
kde_features = ["Pkt_Per_Src","Burst_Rate","Inter_Arrival_Time","BoxCox_Length"]
for f in kde_features:
    if f in df_test_sample.columns:
        plt.figure(figsize=(7,4))
        sns.kdeplot(df_test_sample.loc[df_test_sample['Consensus_Anomaly']==0, f], label="Normal (consensus)", fill=True)
        sns.kdeplot(df_test_sample.loc[df_test_sample['Consensus_Anomaly']==1, f], label="Consensus Anom", fill=True)
        plt.title(f"KDE Test Sample: {f}")
        plt.legend()
        plt.show()

# Confusion matrix for RF if available
if rf is not None:
    labels = list(rf.classes_)
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=labels, yticklabels=labels, cmap="Blues")
    plt.title("RandomForest confusion matrix (test sample)")
    plt.ylabel("True")
    plt.xlabel("Predicted")
    plt.show()

# -------------------------
# 14. Persist summary to disk
# -------------------------
out = {
    "summary": summary,
    "rf_feature_importances": feat_imp.to_dict() if rf is not None else {},
    "top_botnets_test": df_test_sample['Pred_Botnet_Profile'].value_counts().head(20).to_dict()
}
with open("detection_summary.json", "w") as fh:
    json.dump(out, fh, indent=2)

print("Done. Summary saved to detection_summary.json")
```
